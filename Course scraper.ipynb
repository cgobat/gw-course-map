```python
import re, requests, networkx as nx, matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs
from pprint import pprint

with open("subjects.txt") as subjList:
    subjects = [subj.lower() for subj in subjList.read().splitlines()]
#subjects = ['astr','csci','math','phys','stat'] #overwrites the list of all subjects so you can limit to subjects of interest
```

```python
masterdict = {} # initialize and empty dictionary
for subject in subjects:
    page = requests.get('http://bulletin.gwu.edu/courses/'+subject+'/') # retrieve the relevant page from the GW Bulletin
    soup = bs(page.text, 'html.parser') # parse the HTML
    courses = soup.find_all('div',{"class":"courseblock"}) # return the course data--we don't care about anything else
    subjdict = {} # initialize an empty dictionary
    for course in courses:
        title = course.find('p',{"class":"courseblocktitle"}).get_text()
        title = title.replace('\xa0',' ')[:title.index(".")] # replace weird character with space. Cut out everything after the .
        desc = course.find('p',{"class":"courseblockdesc"}).get_text()
        desc = desc.replace('\xa0',' ').split(": ") # can't slice by index like before because there might be multiple : in the description, so split instead
        subjdict[title] = desc[-1] # after splitting over :, we want the last part, which is the prereq list
    masterdict.update(subjdict) # add this subject's dictionary to the main one
    print("Loaded",subject.upper(),"courses successfully.") # report progress
pprint(masterdict) # pretty print for manual verification
```

```python
tree = nx.DiGraph() # create a direction graph object using NetworkX
#tree.add_nodes_from(masterdict.keys())

for prereq in masterdict:
    for dependent in masterdict:
        if prereq in masterdict[dependent]: # check to see if the course in question appears as a prereq in any other courses' descriptions
            tree.add_edge(prereq, dependent) # if so, create a directional relationship between the two. Nodes are created automatically.
```

```python
fig = plt.figure(figsize=(50,40),dpi=300) # make it BIG
nx.draw(tree, with_labels=True, node_size=100, edge_color="grey") # render the network. This may take some time.
fig.savefig("network.png",bbox_inches="tight") # save image to disk so you can ZOOM
```

```python
print("This graph comprises",tree.number_of_nodes(),"nodes with",tree.number_of_edges(),"connections out of a total of",len(masterdict),"courses offered.")
nx.write_graphml(tree, 'tree_graph.xml') # export to GraphML file, which can be imported into Cytoscape, for example
```
